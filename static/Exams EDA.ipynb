```python
import pandas as pd
import numpy as np
```

```python
from urllib.request import urlopen
from json import loads


url = f"https://educationdata.urban.org/api/v1/schools/crdc/ap-exams/2017/race/sex/"
response = urlopen(url)
data = loads(response.read())
ap_df = pd.DataFrame(data['results'], 
              columns=['year', 'fips', 'race', 'sex', 'students_AP_exam_none', 'students_AP_pass_none', 'students_AP_exam_oneormore', 'students_AP_pass_oneormore', 'students_AP_exam_all', 'students_AP_pass_all'])
```

```python
races = {1:'White', 2:'Black', 3:'Hispanic', 4:'Asian', 5:'American Indian or Alaska Native', 
         6:'Native Hawaiian or other Pacific Islander', 7:'Two or more races'}

ap_df = ap_df[ap_df['race'] != 99]
ap_df = ap_df.replace({"race": races})
ap_df.head()
```

```python
ap_df.replace([-2, -1, -3], np.nan, inplace=True) # invalid inputs

# other cols are all invalid so removed
final_ap_df = ap_df.groupby('race')[['students_AP_exam_none',
                                  'students_AP_exam_oneormore']].sum()

final_ap_df = final_ap_df.reset_index()
final_ap_df.columns = ['Race', 'y0', 'y1']
final_ap_df
```

```python
final_ap_df.to_csv('final_ap_df.csv', index=False)
# final_ap_df0.to_csv('final_ap_df0.csv', index=False)
```
